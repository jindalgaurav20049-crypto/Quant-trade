# Momentum Factor Analysis

This notebook analyzes the momentum factor for Indian markets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from data.fetchers.yfinance_fetcher import fetch_multiple
from strategies.momentum.momentum_factor_portfolio import MomentumFactorPortfolio
```

```python
# Fetch data
tickers = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'ICICIBANK.NS', 'INFY.NS']
prices = fetch_multiple(tickers, period='2y')
```

```python
# Build price matrix
close_prices = pd.DataFrame({t: df['Close'] for t, df in prices.items()}).dropna()
close_prices.head()
```

```python
# Run momentum strategy
strategy = MomentumFactorPortfolio(
    universe_tickers=tickers,
    lookback_months=12,
    skip_months=1,
    top_n=3
)

signals = strategy.generate_signals(close_prices)
signals
```

```python
# Plot momentum scores
momentum = strategy.compute_momentum_signal(close_prices)
momentum.plot(kind='bar')
plt.title('12-1 Momentum Scores')
plt.ylabel('Momentum')
plt.show()
```
